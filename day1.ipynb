```python
import torch

#张量

tensor_A = torch.arange(12).reshape(3,4)
tensor_B = torch.arange(12,24).reshape(3,4)

tensor_A,tensor_B
```

```python
# cat 使用cat进行拼接
cat_0 = torch.cat((tensor_A,tensor_B),dim=0) #按行拼接
cat_1 = torch.cat((tensor_A,tensor_B),dim=1) #按列拼接
cat_0,cat_1
```

```python
# 广播机制
board_A = torch.arange(12).reshape(3,4)
board_B = torch.arange(12,16)
board_A+board_B
```

```python
import os
import pandas as pd

os.makedirs(os.path.join('..','data'),exist_ok=True)
data_file = os.path.join('..','data','house_tiny.csv')
with open(data_file, 'w') as f:
    f.write('NumRooms,Alley,Price\n')
    f.write('NA,Pave,127500\n')
    f.write('2,NA,106000\n')
    f.write('4,NA,178100\n')
    f.write('NA,NA,140000\n')
```

```python
data = pd.read_csv(data_file)
data
```

```python
inputs,outputs = data.iloc[:,0:2],data.iloc[:,2:3]
inputs,outputs
```

```python
inputs = inputs.fillna(inputs.mean())
inputs
```

```python
inputs = pd.get_dummies(inputs,dummy_na=True)
inputs
```

```python
inputs_tensor,outputs_tensor = torch.tensor(inputs.values),torch.tensor(outputs.values)
inputs_tensor,outputs_tensor
```

```python
all = torch.cat((inputs_tensor,outputs_tensor),dim=1)
all
```
